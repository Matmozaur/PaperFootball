```python
    def get_positions(x, y):
        position1 = [0, 0]
        position2 = [0, 0]
        position1[1] = x
        if y % 4 == 0:
            position2[1] = x+1
            position1[0] = int(y/4)
            position2[0] = int(y/4)
        elif y % 4 == 1:
            position2[1] = x
            position1[0] = int(y/4)
            position2[0] = int(y/4+1)
        elif y % 4 == 2:
            position2[1] = x+1
            position1[0] = int(y/4)
            position2[0] = int(y/4)+1
        elif y % 4 == 3:
            position2[1] = x+1
            position1[0] = int(y/4)+1
            position2[0] = int(y/4)
        return position1, position2

    def get_all_lines(board):
        lines = []
        for i in range(48):
            for j in range(8):
                if board[i, j] == 1:
#                     print(j,i)
                    lines.append(get_positions(j, i))
        return lines
```
